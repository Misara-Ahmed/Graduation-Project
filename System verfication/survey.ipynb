```python
import matplotlib.pyplot as plt

# Data
total_responses = 38
no_responses = 9
yes_responses = total_responses - no_responses

# Pie chart data
# labels = 'Want to buy our device', 'Don\'t want to buy the device'
sizes = [yes_responses, no_responses]
colors = ['lightblue', '#F6ABD4']
explode = (0.1, 0)  # explode the 'Yes' slice

# Plotting the pie chart
fig1, ax1 = plt.subplots(figsize=(4,2))
ax1.pie(sizes, explode=explode, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90,textprops={'fontsize': 12, },labeldistance=0.8)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
# plt.xticks(fontsize=14)  # Increase label font size
# plt.yticks(fontsize=14) 
# Display the plot with modified title
# plt.title('Survey Responses')
plt.show()
```
